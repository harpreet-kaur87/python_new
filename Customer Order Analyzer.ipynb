#### 🛒 Customer Order Analyzer

### 🧩 Problem Statement

Given a dataset of customer orders, where each customer has multiple order amounts, write a function to analyze customer spending based on the following rules:

* Exclude customers if any order amount is less than **100**
* Calculate total spending for each valid customer
* Apply a **10% bonus** if total spending exceeds **5000**
* Stop processing further customers if total exceeds **10000**
* Categorize customers as:

  * **Premium** (≥ 8000)
  * **Gold** (5000–7999)
  * **Silver** (< 5000)
* Identify the **highest spending customer(s)**

### 📤 Output

Return:

* A dictionary with customer spending and category
* A list of highest spending customer(s)

```python
orders = { "Alice": [1200, 1500, 2000], 
          "Bob": [500, 700, 50],
          "Charlie": [3000, 2500, 1000], 
          "David": [800, 900, 950], 
          "Eva": [4000, 3000, 2000] }


def customer_order_analyzer(orders):
    
    order_status = {}
    
    for k,v in orders.items():
        if any(amt < 100 for amt in v):
            continue
            
        total_amount = sum(v)
        
        if total_amount > 5000:
            total_amount = total_amount * 1.1
            
        if total_amount > 10000:
            break
            
        if total_amount >= 8000:
            status = 'Premium'
        elif total_amount >= 5000:
            status = 'Gold'
        else:
            status = 'Silver'
            
        order_status[k] = (total_amount,status)
        
    highest_total = max(order_status.values(), key=lambda x:x[0])
    highest_total_spenders = [k for k,v in order_status.items() if v[0] == highest_total[0]]
        
    return order_status, highest_total_spenders

customer_order_analyzer(orders)
```
