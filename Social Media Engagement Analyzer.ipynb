### 📖 Problem Statement

This project builds a simple Social Media Engagement Analyzer using Python. It calculates a weighted engagement score for each post based on likes, comments, and shares. Posts with engagement above 200 are included in the results, and processing stops if any post exceeds a score of 500. The final output is returned as a dictionary mapping post IDs to their engagement scores.

Task :

Calculate engagement score (likes + comments2 + shares3)  
Filter posts above 200 score  
Stop if engagement > 500  
Return engagement dictionary  


```python
posts = [
    {"id": 1, "likes": 700, "comments": 20, "shares": 10},
    {"id": 2, "likes": 300, "comments": 50, "shares": 40},
    {"id": 3, "likes": 50, "comments": 5, "shares": 2}
]

def social_media_analyzer(lst):

    engagement = {}

    for post in lst:
        score = post['likes']+post['comments']+post['shares']
        if score > 500:
            continue
        elif score > 200:
            engagement[post['id']] = score
    return engagement

result = social_media_analyzer(posts)
result
```
