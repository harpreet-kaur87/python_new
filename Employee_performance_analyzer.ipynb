#### Problem Statement:
Given a list of employees (ID, Name, Department) and a dictionary of performance scores mapped by Employee ID, calculate the average score for each employee. Return a new dictionary where the keys are employee names and the values are their average performance scores.

```python
employees = [
    ("E01", "Alice", "Sales"),
    ("E02", "Bob", "IT"),
    ("E03", "Charlie", "Sales"),
    ("E04", "David", "HR"),
]

performance_scores = {
    "E01": [78, 85, 90],
    "E02": [88, 92],
    "E03": [95, 91, 89],
    "E04": [60, 65, 70]
}

def performance_analyzer(employees,performance_scores):
    
    result = {}
    final_result = {}
    
    for emp_id,name,dept in employees:
        result[emp_id] = name
    
    for k,v in performance_scores.items():
        avg_score = sum(v)/len(v)
        name = result[k]
        final_result[name] = avg_score
    return final_result
        
performance_analyzer(employees,performance_scores)
```
