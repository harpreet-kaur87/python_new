#### Online Course Analytics System

#### Description
A Python-based mini project to analyze course completion data and generate insights such as average completion rate, top-performing course, and user engagement.

🔹 Features  
Average completion per course  
Top course by completion rate  
Users enrolled in multiple courses  
Courses with low completion (<50%)  

```python
courses = [
    ("U1", "Python", 80),
    ("U2", "Python", 60),
    ("U1", "SQL", 40),
    ("U3", "SQL", 30),
    ("U2", "ML", 90),
]

def online_course_platform(lst):

    result = {}
    user_count = {}
    users_enrolled_in_multiple_courses = []
    courses_with_less_completion = []

    for user,course,marks in lst:

        if course not in result:
            result[course] = []

        result[course].append(marks)

        if user in user_count:
            user_count[user] += 1
        else:
            user_count[user] = 1

    for k,v in user_count.items():
            if v > 1:
                users_enrolled_in_multiple_courses.append(k)
                
    avg_completion_rate = {k:sum(v)/len(v) for k,v in result.items()}

    top_course_by_completion = max(avg_completion_rate.items(), key=lambda x:x[1])


    for k,v in avg_completion_rate.items():
        if v < 50:
            courses_with_less_completion.append(k)            

    return {
          "Avg Completion Rate per Course" : avg_completion_rate,
          "Top Course by Completion" : top_course_by_completion,
          "User Enrolled in more than 1 Course" : users_enrolled_in_multiple_courses,
          "Courses with Completion <50%" : courses_with_less_completion
    }

ans = online_course_platform(courses)
ans
```
