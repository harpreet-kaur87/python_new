#### 📄 Problem Statement: Hotel Booking System

🎯 Objective

The objective of this project is to build a Python-based analytics system that processes hotel booking data and generates meaningful insights such as total revenue, revenue distribution across room types, and booking trends.

⚙️ Key Requirements
The system should:

1. Calculate total revenue  
Consider only confirmed bookings  
Ignore cancelled bookings  
2. Track unique room types  
Include all room types, regardless of booking status  
3. Compute revenue by room type  
Aggregate revenue for each room category  
4. Count bookings per room type  
Only include confirmed bookings  
5. Identify highest revenue-generating room type  
Determine which room category contributes the most revenue  
6. Identify most frequently booked room type  
Based on number of confirmed bookings  

🧠 Expected Outcome
The system should return a structured output (dictionary) containing:  

Total revenue  
List of unique room types  
Revenue distribution per room type  
Booking count per room type  
Highest revenue-generating room type  
Most frequently booked room type  

```python
bookings = [
    ("B1", "Deluxe", 5000, "Confirmed"),
    ("B2", "Suite", 7000, "Cancelled"),
    ("B3", "Standard", 2000, "Confirmed"),
    ("B4", "Standard", 2000, "Cancelled"),
    ("B5", "Suite", 8000, "Confirmed"),
    ("B6", "Suite", 3000, "Confirmed")
]

def hotel_booking_system(lst):
    
    total = 0
    unique_room_types = []
    revenue_by_room = {}
    booking_per_room = {}

    for booking_id, room_type, amt, status in lst:
        
        if room_type not in unique_room_types:
            unique_room_types.append(room_type)
            
        if room_type not in revenue_by_room:
            revenue_by_room[room_type] = 0
            
        if room_type not in booking_per_room:
            booking_per_room[room_type] = 0
    
        if status.lower() == 'cancelled':
            continue
        
        revenue_by_room[room_type] += amt
        
        booking_per_room[room_type] += 1
        
    total = sum(revenue_by_room.values())
        
    highest_revenue_room_type = max(revenue_by_room.items(), key=lambda x:x[1])
    most_booked_room_type = max(booking_per_room.items(), key=lambda x:x[1])
        
    return {"total" : total,
            "unique room types" : unique_room_types,
            "revenue by room type" : revenue_by_room,
            "highest revenue room type" : highest_revenue_room_type,
            "most booked room type" :most_booked_room_type
           }
hotel_booking_system(bookings)
```
