#### 📄 Hospital Revenue Analytics System

Analyze hospital patient billing data to generate key revenue insights.

🎯 Objective

Build a Python-based system to process patient records and compute revenue metrics across treatments and payment modes.

⚙️ Key Features  
Calculate total revenue  
Identify high-value transactions (≥ 10,000)  
Compute OPD revenue (< 10,000)  
Analyze surgery revenue and patient count  
Break down revenue by payment mode (Cash/Insurance)  
Extract list of high-value patients  

🧠 Expected Output Structure  
{
    "total_revenue": ...,  
    "high_value_revenue": {...},  
    "opd_revenue": ...,  
    "surgery_revenue": ...,  
    "surgery_count": ...,  
    "payment_mode_revenue": {...},  
    "high_value_patients": [...]
}

```python
patients = [
    ("P1", "Surgery", 50000, "Cash"),
    ("P2", "Consultation", 2000, "Insurance"),
    ("P3", "Therapy", 15000, "Cash"),
    ("P4", "Surgery", 70000, "Insurance"),
    ("P5", "Consultation", 3000, "Cash"),
    ("P6", "Therapy", 12000, "Insurance"),
    ("P7", "Surgery", 45000, "Cash"),
    ("P8", "Consultation", 8000, "Cash")
]

def hospital_revenue_system(lst):
    
    total_revenue = 0
    high_value_revenue = 0
    opd_revenue = 0
    no_of_surgery_patients = 0
    revenue_by_treatment = {}
    payment_mode = {}
    high_value_treatment_list = []
    
    for patient,condition,amt,payment_method in lst:
        
        total_revenue += amt
        
        if amt >= 10000:
            high_value_revenue += amt
            
            high_value_treatment_list.append((patient,condition,amt))
            
        if amt < 10000:
            opd_revenue += amt
            
        if condition.lower() == 'surgery':
            no_of_surgery_patients += 1
            
        if condition not in revenue_by_treatment:
            revenue_by_treatment[condition] = 0
        
        revenue_by_treatment[condition] += amt
        
        if payment_method not in payment_mode:
            payment_mode[payment_method] = 0
            
        payment_mode[payment_method] += amt
        
        surgery_total_revenue = {(k,v) for k,v in revenue_by_treatment.items() if k.lower() == 'surgery'}
    
        
    return {
            "total_revenue" : total_revenue, 
            "high value revenue" : high_value_revenue, 
            "opd revenue" : opd_revenue, 
            "surgery revenue" : surgery_total_revenue,
            "surgery count" : no_of_surgery_patients,
            "revenue by payment mode" : payment_mode,
            "high value patients" : high_value_treatment_list
    }
        
hospital_revenue_system(patients)
```
