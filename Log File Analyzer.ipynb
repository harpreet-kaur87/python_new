###  📄 Log File Analyzer

🧩 Problem Statement

You are given a list of system log entries. Each log is a string containing information about user activity in the following format:

📌 Requirements

1. **Parse Logs**
* Extract User ID, Action, Status, and Time from each log entry

2. **Track Failed Attempts**
* Count failed login attempts per user
* If a user has **2 or more failed logins**, mark them as `"Blocked"`

3. **Track Successful Actions**
* Count total successful actions per user

4. **Ignore Blocked Users**
* Once a user is marked `"Blocked"`, ignore all further activities of that user

📤 Output

Return a dictionary where:  
Each key is a User ID  
Each value is another dictionary containing:  
status: "Active" or "Blocked"  
success_count: total number of successful actions  

```python
logs = [ "U101 | LOGIN | SUCCESS | 09:00", 
        "U102 | LOGIN | FAILED | 09:05", 
        "U101 | PURCHASE | SUCCESS | 09:10", 
        "U103 | LOGIN | SUCCESS | 09:15", 
        "U102 | LOGIN | FAILED | 09:20", 
        "U102 | LOGIN | SUCCESS | 09:25" ]

users = {}

for log in logs:
        user,action,status,time = log.split('|')
        user = user.strip()
        action = action.strip().lower()
        status = status.strip().lower()
        time = time.strip()
        
        if user not in users:
            users[user] = {
            "failed" : 0,
            "success" : 0,
            "blocked" : False}
            
        if users[user]["blocked"]:
            continue
            
        if action == 'login' and status == 'failed':
            users[user]['failed'] += 1
            
            if users[user]['failed'] >= 2:
                users[user]['blocked'] = True
                
        elif status == 'success':
            users[user]['success'] += 1
            
result = {}

for user,data in users.items():
    result[user] = {'status' : 'Blocked' if data['blocked'] else 'Active',
                    'success_count' : data['success']
                   }
        
print(result)       
```
