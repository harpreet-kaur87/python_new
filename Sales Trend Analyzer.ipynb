#### 📊 Sales Trend Analyzer

Problem Statement :

You are given a dataset representing monthly sales figures for different regions. Each region contains a list of sales values.
Your task is to analyze this data and generate a summary based on the following business rules:

### 📌 Requirements
1. **Skip Low-Performing Regions**
* If any monthly sales value in a region is less than **16,000**, exclude that entire region from analysis.
2. **Calculate Total Sales**
* For each valid region, calculate the **total sales** by summing all monthly values.
3. **Early Termination Condition**
* If the total sales of any region exceed **100,000**, stop processing further regions immediately.
4. **Identify Top Performer**
* Determine the region(s) with the **highest total sales** among the processed regions.

Return:
1. A dictionary with **total sales per valid region**
2. A list of **highest performing region(s)**

```python
sales = {
    "North": [20000, 25000, 30000],
    "South": [15000, 18000, 17000],
    "East": [22000, 21000, 23000],
    "West": [30000, 35000, 40000]
}

def sales_trend_analyzer(sales):

    final_summary = {}

    for k,v in sales.items():
        if any(sale < 16000 for sale in v):
            continue
            
        total = sum(v)
        
        if total > 100000:
            break
        final_summary[k] = total
        
    highest_value = max(final_summary.values())
    highest_performing_region = [k for k,v in final_summary.items() if v == highest_value]

    return f'{final_summary}, Highest Performing Region - {highest_performing_region}'

print(sales_trend_analyzer(sales))
```
