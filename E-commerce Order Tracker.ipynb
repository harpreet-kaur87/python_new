#### 📦 E-commerce Order Tracker

📌 Problem Statement

You are given a list of order logs representing user transactions in an e-commerce system. Each log contains:

user_id  
event_type (always "ORDER" in this case)  
status ("SUCCESS" or "FAILED")  
amount (order value)  

🎯 Objective

Build a system to:  

✅ Track total successful order amount per user  
✅ Count failed orders per user  
✅ Mark a user as "Blocked" if they have 2 or more failed orders  
✅ Ignore all further logs once a user is blocked  

```python
orders = [
    ("U1", "ORDER", "SUCCESS", 250),
    ("U2", "ORDER", "FAILED", 300),
    ("U1", "ORDER", "SUCCESS", 150),
    ("U2", "ORDER", "SUCCESS", 200),
    ("U3", "ORDER", "SUCCESS", 500),
    ("U2", "ORDER", "FAILED", 100),
    ("U2", "ORDER", "FAILED", 50)
]

users = {}

for user,order,status,amt in orders:
    if user not in users:
        users[user] = {
        "total_amt" : 0,
        "failed_orders" : 0,
        "blocked" : False
    }
        
    if users[user]["blocked"]:
        continue

    if status.lower() == 'success':
        users[user]["total_amt"] += amt

    elif status.lower() == 'failed':
        users[user]["failed_orders"] += 1

        if users[user]["failed_orders"] >= 2:
            users[user]["blocked"] = True

result = {}
                
for key,value in users.items():
    result[key] = {
        "status" : "Blocked" if value["blocked"] else "Active",
        "total_spent" : value["total_amt"]
    }
    
for k,v in result.items():
    print({k:v})
```
