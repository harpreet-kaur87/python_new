#### Problem Statement:

🎯 Task

Calculate average score per student.

Classify:

"Distinction" → average ≥ 90

"Pass" → average ≥ 50 and < 90

"Fail" → average < 50

Remove any student who has a score below 40 in any subject.

Return a list sorted by average descending:
[(name, average, grade)]


```python
students = [
    ["Alice", [88, 92, 79]],
    ["Bob", [100, 65, 70]],
    ["Charlie", [35, 55, 60]],
    ["David", [99, 98, 97]]
]

def calculate_score(lst):

        result = []

        for name,marks in lst:
            remove_student = False
            for mark in marks:
                if mark < 40:
                    remove_student = True
                    break
            if remove_student:
                continue
            avg_score = round(sum(marks)/len(marks),0)

            if avg_score >= 90:
                grade = 'Distinction'
            elif avg_score >= 50:
                grade = 'Pass'
            else:
                grade = 'Fail'

            result.append((name,avg_score,grade))
        result = sorted(result, key=lambda x:x[1], reverse=True)
        return result
    
calculate_score(students)
```
