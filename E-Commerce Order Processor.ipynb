ðŸ“Œ Problem Statement

In an e-commerce system, each order contains an order ID, customer name, a list of item prices, and the order status. The goal of this project is to process a collection of such orders and generate a summary of valid transactions.

ðŸŽ¯ Objectives

Build a Python function that:

Calculates the total order value by summing all item prices in each order.

Ignores cancelled orders and processes only completed ones.

Tracks unique customers using a set data structure.

Stops processing further orders if any single orderâ€™s total value exceeds â‚¹4000.

Returns a dictionary in the format:
{OrderID: TotalOrderValue}

```python
orders = [
    ("O101", "Alice", [1200, 800, 400], "Completed"),
    ("O102", "Bob", [500, 700], "Cancelled"),
    ("O103", "Charlie", [1500, 2000], "Completed"),
    ("O104", "Alice", [3000], "Completed"),
]

def process_order(lst):
    
    records = {}
    unique_customers = set()
    
    for order_id,name,value,status in lst:
        if status.lower() == 'cancelled':
            continue
        total = sum(value)
        if total > 4000:
            break
        records[order_id] = total
        unique_customers.add(name)
    print(f'Unique Customers - {unique_customers}')
    
    return records
            

process_order(orders)
```
