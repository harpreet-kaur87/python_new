In a retail inventory system, each product is associated with stock quantities stored across multiple warehouse locations. Additionally, certain products may be marked as discontinued and should not be included in active stock evaluation.
The objective of this project is to build a Python function that analyzes inventory data and identifies products that require restocking based on defined business rules.

ðŸŽ¯ Objectives
Develop a function that:
1. Calculates the total stock per product by summing quantities across all warehouse locations.
2. Skips discontinued products from processing.
3. Identifies products with total stock below 20 units.
4. Stops processing immediately if any productâ€™s total stock exceeds 100 units.
5. Returns a list of low-stock products.

```python
inventory = {
    "Laptop": [10, 5, 0],
    "Mobile": [50, 40, 30],
    "Tablet": [0, 0, 0],
    "Monitor": [15, 10, 5]
}
discontinued = {"Tablet"}

def track_inventory_stock(inventory,discontinued):
    
    low_stock_list = []
    
    for k,v in inventory.items():
        if k in discontinued:
            continue
        product_stock_total = sum(v)
        if product_stock_total > 100:
            break
        if product_stock_total < 20:
            low_stock_list.append(k)
        
    return low_stock_list

        
result = track_inventory_stock(inventory,discontinued)
print(f'Low stock items - {result}')
```
